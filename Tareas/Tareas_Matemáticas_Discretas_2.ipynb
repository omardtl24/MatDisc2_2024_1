# Tarea 1

Determinar si la operación $⋆$ es una operación binaria sobre el conjunto dado $S$:

a. $S = \mathbb{R}$ , $a⋆b = \dfrac{a}{a^2 + b^2}$

b. $S = \{1,-2,3,2,-4\}$ , $a⋆b = |b|$

c. $S = \mathbb{R}$ , $a⋆b = a$




## Desarrollo

### a. $S = \mathbb{R}$ , $a⋆b = \dfrac{a}{a^2 + b^2}$

**Contraejemplo**

Vamos a definir dos valores $a,b \in \mathbb{R}$

$$0 \in \mathbb{R} \,\, \wedge  \,\, a=b=0 \,\, \Longrightarrow a,b \in \mathbb{R}$$

Aplicamos nuestra operación $a⋆b$

$$a⋆b = 0⋆0 = \dfrac{0}{0^2 + 0^2} = \dfrac{0}{0} \notin \mathbb{R} \Longrightarrow a⋆b \notin \mathbb{R}$$

Encontramos una pareja $(a,b) \in \mathbb{R} \times \mathbb{R}$ tal que $a⋆b \notin \mathbb{R}$

$$\therefore ⋆ \text{ NO es una operación binaria}$$


### b. $S = \{1,-2,3,2,-4\}$ , $a⋆b = |b|$

Dado que tenemos un conjunto finito, podemos realizar todas las operaciones sin problema. Por practicidad, vamos a organizarlo en una tabla de $n \times n$ donde $n = |S|$

Siendo $a$ el valor de la fila y $b$ el valor de la columna, la siguiente tabla tiene los resultados de la operación $a⋆b$

\begin{array}{|r|c|c|c|c|c|}
    \hline
     S \times S & \textbf{1} & \textbf{-2} & \textbf{3} & \textbf{2} & \textbf{-4} \\
    \hline
    \textbf{1} & 1 & 2 & 3 & 2 & 4 \\
    \hline
    \textbf{-2} & 1 & 2 & 3 & 2 & 4 \\
    \hline
    \textbf{3} & 1 & 2 & 3 & 2 & 4 \\
    \hline
    \textbf{2} & 1 & 2 & 3 & 2 & 4 \\
    \hline
    \textbf{-4} & 1 & 2 & 3 & 2 & 4 \\
    \hline
\end{array}

Podemos denotar que en la quinta columna (Asociada a $b=-4$) obtenemos un resultado que no está en el conjunto original $S$

$$a⋆ -4 = 4 \notin S \,\,\,\, \forall a \in S$$

$$\therefore ⋆ \text{ NO es una operación binaria}$$

### c. $S = \mathbb{R}$ , $a⋆b = a$

No es necesario ser muy metódicos ni formales para denotar que si se tiene por precondición que los operandos $a$ y $b$ pertenecen al conjunto $S$ y el resultado de la operación es uno de esos mismos operandos (En este caso $a$), se puede concluir que $a⋆b$ tambien pertenece al conjunto $S$

$$a,b \in S \,\, \wedge \,\, a⋆b = a \Longrightarrow a⋆b \in S$$

$$\therefore ⋆ \text{ SI es una operación binaria}$$


# Tarea 2

Demuestre que $\mathbb{R}−\{1\}$ bajo la operación $a∘b=a+b−ab$ es un grupo.

## Desarollo

Para demostrar que $(\mathbb{R}−\{1\} , ∘)$ es un grupo, debemos determinar que $∘$ es una operación binaria sobre el conjunto, que es una operación asociativa, que existe un elemento identidad y que cada elemento de un conjunto tiene un inverso.

### ¿Es $∘$ una operación binaria?

Para concluir $∘$ es una operación binaria, necesitamos demostrar que:

$$a,b \in \mathbb{R}−\{1\} \Longrightarrow a ∘ b \in \mathbb{R}−\{1\}$$

En primer lugar, vamos a aprovechar las propiedades de cerradura de la suma $(+)$ y el producto $(\cdot)$ sobre los números reales $(\mathbb{R})$. Estas propiedades son las siguientes:

$$a,b \in \mathbb{R} \Longrightarrow a + b \in \mathbb{R}$$
$$a,b \in \mathbb{R} \Longrightarrow ab \in \mathbb{R}$$

Adicionalmente, podemos escribir nuestro conjunto $\mathbb{R}−\{1\}$ utlilizando la definición de diferencia entre conjuntos.

$$\mathbb{R}−\{1\} = \{x: x \in \mathbb{R} \wedge x \notin \{1\}\} = \{x: x \in \mathbb{R} \wedge x \neq 1\}$$


Podemos denotar en la definición del conjunto, que podemos analizar que el punto clave se encuentra en la *exclusión* del numero 1 del conjunto de los reales. Por lo que el siguiente paso será realizar un análisis por **contradicción**.

Suponemos que:

$$\exists a,b \in \mathbb{R}−\{1\}\,\, (a ∘ b = 1 )$$

$$\exists a,b \neq 1 \,\, (a ∘ b = 1 )$$

Realizamos nuestro desarrollo algebraico:

$$a ∘ b = 1 \Longrightarrow a+b-ab =1$$

\begin{array}{ccccc}
    a-ab = 1-b & & \lor & &b-ab = 1-a \\
    a(1-b) = 1-b &&\lor&& b(1-a) = 1-a \\
    \text{Dado que b}\neq 1 &&\lor&& \text{Dado que a}\neq 1 \\
    a = \dfrac{1-b}{1-b} &&\lor&& b = \dfrac{1-a}{1-a} \\
    a = 1 &&\lor&& b = 1
\end{array}

Dado que obtuvimos que $a = 1 \lor b = 1$ y nuestra suposición era que $a \neq 1 \wedge b \neq 1$, hemos llegado a una **CONTRADICCIÓN**

Por consiguiente, queda demostrado que $a,b \in \mathbb{R}−\{1\} \Longrightarrow a ∘ b \in \mathbb{R}−\{1\}$

$$\therefore ∘ \text{ SI es una operación binaria}$$

### ¿Es $∘$ una operación asociativa?

Para concluir $∘$ es una operación asociativa, necesitamos demostrar que:

$$a,b,c \in \mathbb{R}−\{1\} \Longrightarrow (a ∘ b) ∘ c = a ∘ (b ∘ c)$$

Realizamos el análisis algebraico

\begin{array}{ccc}
    (a ∘ b) ∘ c & = &  (a+b - ab) ∘ c & & \\
                & = &  a+b - ab + c - (a+b - ab)c & & \\
                & = &  a+b - ab + c - (ac+bc - abc) & & \\
                & = &  a + b - ab + c - ac - bc + abc & & \\
                & = &  a + b + c - bc - ac - ab  + abc & & \\
                & = &  a + (b + c - bc) - a(b + c  - bc) & & \\
                & = &  a ∘ (b + c - bc) & = & a ∘ (b ∘ c)\\
\end{array}

Queda demostrado que $a,b,c \in \mathbb{R}−\{1\} \Longrightarrow (a ∘ b) ∘ c = a ∘ (b ∘ c)$

$$\therefore ∘ \text{ SI es una operación asociativa}$$

### ¿Existe un elemento identidad $(e)$?

Para determinar si existe un elemento identidad, debemos encontrar un elemento que cumpla la siguiente propiedad:

$$\forall_{a \in \mathbb{R}−\{1\}} \exists_{e \in \mathbb{R}−\{1\}} \,\, (a ∘ e = e ∘ a = a)$$

Planteamos el desarrollo algebraico para hallar el elemento identidad

\begin{array}{ccccccc}
    a ∘ e = a & \Longrightarrow &  a + e - ae = a & & & & e ∘ a = a & \Longrightarrow &  e + a - ea = a & &\\
              & \Longrightarrow &  e - ae = a - a& & & & & \Longrightarrow &  e - ea = a - a& &\\
              & \Longrightarrow &  e - ae = 0 & & & &  & \Longrightarrow &  e - ea = 0 & &\\
              & \Longrightarrow &  e(1-a) = 0 & & & & & \Longrightarrow &  e(1-a) = 0 & &\\
              &                 & \text{Dado que a}\neq 1 & & & & & & \text{Dado que a}\neq 1 & &\\
              & \Longrightarrow &  e = 0 & & & & & \Longrightarrow &  e = 0 & &\\
\end{array}

Encontramos que el elemento $e = 0$ es aquel que cumple que $a ∘ 0 = 0 ∘ a = a \,\,\, \forall_{a \in \mathbb{R}−\{1\}}$

$$\therefore \text{ SI existe un elemento identidad} \,\, (e = 0)$$

### ¿Existen los elementos inversos?

Para determinar si existen los elementos inversos, debemos encontrar aquellos elementos que satisfacen:

$$\forall_{a \in \mathbb{R}−\{1\}} \exists_{h \in \mathbb{R}−\{1\}} \,\, (a ∘ h = h ∘ a = 0)$$

Planteamos el desarrollo algebraico para hallar el elemento identidad

\begin{array}{ccccccc}
    a ∘ h = 0 & \Longrightarrow &  a + h - ah = 0 & & & & h ∘ a = 0 & \Longrightarrow &  h + a - ha = 0 & &\\
              & \Longrightarrow &  h - ah = -a & & & & & \Longrightarrow &  h - ha = -a& &\\
              & \Longrightarrow &  h(1-a) = -a & & & &  & \Longrightarrow &   h(1-a) = -a & &\\
              & \Longrightarrow &  h(a-1) = a & & & & & \Longrightarrow &  h(a-1) = a & &\\
              &                 & \text{Dado que a}\neq 1 & & & & & & \text{Dado que a}\neq 1 & &\\
              & \Longrightarrow &  h = \dfrac{a}{a-1} & & & & & \Longrightarrow &  h = \dfrac{a}{a-1} & &\\
\end{array}

Encontramos que el elemento $h = \dfrac{a}{a-1}$ es aquel que cumple $\forall_{a \in \mathbb{R}−\{1\}} \exists_{h \in \mathbb{R}−\{1\}} \,\, (a ∘ h = h ∘ a = 0)$

$$\therefore \text{ SI existen los elementos inversos} \,\, \left( h = \dfrac{a}{a-1} \,\,\, \forall_{a \in \mathbb{R}−\{1\}}\right)$$

### Finalmente ...

Dado que identificamos que $(\mathbb{R}−\{1\} , ∘)$ cuenta con una operación básica que es asociable; y que cuenta con un elemento identidad y elementos inversos; podemos concluir que el conjunto dado (sobre la operación $∘$) es un **GRUPO**

$$\therefore (\mathbb{R}−\{1\} , ∘) \text{ SI es un grupo} $$
