# Desarrollo

### 1. ¿Cuál es el residuo de $-33$ al dividirlo por $7$?

Para obtener el residuo de $-33$ al dividirlo por $7$, debemos acogernos a la definición dada por el algoritmo de la división con residuo, el cual esta dado por la siguiente expresión:

$$a = qb+r \quad \text{donde} \quad 0 \leq r < b$$

En nuestro caso particular, sabemos que $a=-33$ y que $b=7$ y que nos interesa hallar es el valor de $r$ que cumple estar entre 0 y $7$ sin incluir este último. Ajustando a nuestra definición, obtenemos lo siguiente:

$$-33 = -5 \cdot 7 + 2$$

Dado que $0 \leq 2 < 7$, podemos afirmar que ese es el valor del residuo de $-33$ al dividirlo por $7$. O en otras palabras, que:

$$\therefore -33 \equiv 2 \,\, (\text{mod  } 7)$$

### 2. ¿Cuál es el residuo de $(34-14 \times 25) \times (23 \times 18 + 87)$ cuando se divide por $5$?

Para determinar este proceso de manera más simple, llevaremos los términos presentes en la ecuación a su término equivalente en $\mathbb{Z}_5$ dada la congruencia módulo $5$:

\begin{array}{lcr}
    34 = 6 \cdot 5 + 4 & \Longrightarrow & 34 \equiv 4 \,\, (\text{mod  } 5) \\
    14 = 2 \cdot 5 + 4 & \Longrightarrow & 14 \equiv 4 \,\, (\text{mod  } 5) \\
    25 = 5 \cdot 5 + 0 & \Longrightarrow & 25 \equiv 0 \,\, (\text{mod  } 5) \\
    23 = 4 \cdot 5 + 3 & \Longrightarrow & 23 \equiv 3 \,\, (\text{mod  } 5) \\
    18 = 3 \cdot 5 + 3 & \Longrightarrow & 18 \equiv 3 \,\, (\text{mod  } 5) \\
    87 = 17 \cdot 5 + 2 & \Longrightarrow & 87 \equiv 2 \,\, (\text{mod  } 5) \\
\end{array}

Ya dados estos términos, procedemos a reescribir nuestra ecuación en base a las equivalencias:

\begin{array}{rcl}
    (34-14 \times 25) \times (23 \times 18 + 87) & \equiv & (4 - 4 \times 0) \times (3 \times 3 + 2) \quad (\text{mod  } 5)\\
                                                 & \equiv & 4 \times 11 \quad (\text{mod  } 5) \\
                                                 & \equiv & 36 \quad (\text{mod  } 5) \\
                                                 & \equiv & 1 \quad (\text{mod  } 5)\\
\end{array}

A partir de este proceso de congruencias módulo 5, podemos concluir que el residuo de la expresión es $1$. O en otras palabras que:

$$\therefore (34-14 \times 25) \times (23 \times 18 + 87) \equiv  1 \,\, (\text{mod  } 5)$$

### 3. ¿Cuál es el residuo de $12 ^ {100}$ cuando se divide por $11$?

Para determinar está expresión de manera más simple, vamos a hallar la congruencia modulo $11$ de la base, para luego si aplicar la potencia requerida:

$$ 12 = 1 \cdot 11 + 1  \Longrightarrow  12 \equiv 1 \,\, (\text{mod  } 11)$$

Reescribimos el problema en base a la congruencia hallada:

 $$12 ^ {100} \equiv 1 ^ {100} \equiv 1 \,\, (\text{mod  } 11)$$

A partir de este proceso de congruencias módulo 11, podemos concluir que el residuo de la expresión es $1$. O en otras palabras que:

$$\therefore 12 ^ {100} \equiv  1 \,\, (\text{mod  } 11)$$
