### 1. Dado $G = \mathbb{Z}_3 \times \mathbb{Z}_2$ y siendo $H$ el subgrupo normal $\{0\} \times \mathbb{Z}_2$

Para una definición un poco más completa, podemos definir nuestro grupo y subgrupo escritos en forma de extensión:

$$G = \mathbb{Z}_3 \times \mathbb{Z}_2 = \left\{ 
 \begin{array}{c}
    (0,0) , (1,0), (2,0), \\
    (0,1) , (1,1), (2,1)
 \end{array} \right\}$$

 $$H = \{0\} \times \mathbb{Z}_2 = \left\{ 
 \begin{array}{c}
    (0,0) , (0,1)
 \end{array} \right\}$$

####    1a) ¿Por qué $H$ es normal en $G$?

Para probar que nuestro subgrupo es normal, vamos a aprovechar la siguiente definición:

$$H \text{  es Normal  } \quad \Longleftrightarrow \quad \forall_{h \in H}\,\, \forall_{g \in G} \,\,\, g  h g^{-1} \, \in H$$

Adicionalmente, vamos a aprovechar que la operación suma módulo n $(\oplus)$ es conmutativa.

$$ a \oplus b = b \oplus a \quad \forall_{a,b \in \mathbb{Z}_n}$$

Suponemos la existencia de un par de elementos independientes que pertenecen a $G$ y a $H$ respectivamente; y a partir de ellos, vamos a denotar si la relación expuesta en la definición de subgrupo normal se cumple.

Dado $(h_1, h_2 ) \in H$ y $(g_1 , g_2) \in G$:

$$
\begin{array}{rclcr}
   (g_1, g_2) \oplus (h_1, h_2) \oplus (g_1^{-1}, g_2^{-1}) & = &(g_1 \oplus h_1 \oplus g_1^{-1}, g_2 \oplus h_2 \oplus g_2^{-1}) & &\text{Aplicamos operación }\oplus \\
                                                            & = &(g_1 \oplus g_1^{-1} \oplus h_1, g_2 \oplus g_2^{-1} \oplus h_2) & &\text{Propiedad Conmutativa } \\
                                                            & = &(e \oplus h_1, e \oplus h_2) & &\text{Propiedad inverso} \\
                                                            & = &(h_1, h_2) \in H & &\text{Propiedad identidad} \\
\end{array}
$$

Podemos concluir:

$$\therefore H \text{  es Normal  }$$


    
####    1b) Encuentra $G/H$ y determina si es cíclico.

Dado que conocemos $H$ y $G$ por extensión, podemos determinar sin mucho problema sus cardinalidades. Adicionalmente, podemos determinar el índice del grupo usando el teorema de Lagrange:

$$|G| = 6 \,\, \wedge \,\, |H| = 2 \,\, \wedge \,\, |G| = [G:H] \cdot |H| \quad \Longrightarrow \quad [G:H] = 3$$

Conociendo el índice, podemos determinar la cardinalidad del conjunto $G/H$ y podemos determinar la cantidad de clases laterales que se deben buscar.

$$|G/H| = [G:H] \,\, \wedge \,\, [G:H] = 3 \quad \Longrightarrow \quad |G/H| = 3$$

$$H \oplus (0,0) = H_{(0,0)} = \{ (0,0) , (0,1) \}$$
$$H \oplus (1,0) = H_{(1,0)} = \{ (1,0) , (1,1) \}$$
$$H \oplus (2,0) = H_{(2,0)} = \{ (2,0) , (2,1) \}$$

Podemos afirmar que el grupo $G/H$ está definido por:

$$G/H = \{H_{(0,0)} \, , \, H_{(1,0)} \, , \, H_{(2,0)}\}$$

Para determinar si el grupo es cíclico, debemos determinar si dispone de un generador.

$$
\begin{array}{rcl}
   \left(H_{(1,0)}\right)^{1} &=& H_{(1,0)} \\
   \left(H_{(1,0)}\right)^{2} &=& H_{(1,0) \oplus (1,0)} = H_{(2,0)} \\
   \left(H_{(1,0)}\right)^{3} &=& H_{(2,0) \oplus (1,0)} = H_{(0,0)} \\
\end{array}
$$

Podemos denotar que a partir de $H_{(1,0)}$ podemos generar todo el grupo. Es decir:

$$ G/H = \langle H_{(1,0)} \rangle$$

Podemos concluir:

$$\therefore G/H = \{H_{(0,0)} \, , \, H_{(1,0)} \, , \, H_{(2,0)}\} \quad \wedge \quad G/H \text{ es cíclico}$$

### 2. Dado $G = (\mathbb{Z},+)$ y $H=\langle 5 \rangle$

Para una definición un poco más completa, podemos definir nuestro grupo y subgrupo escritos en forma de extensión:

$$G = (\mathbb{Z},+) = \{...,-4, -3 , -2 -1, 0, 1 ,2 ,3 ,4 , ...\}$$

$$H=\langle 5 \rangle= \{...,-20, -15 , -10 -5, 0, 5 ,10 ,15 , 20 , ...\} = 5\mathbb{Z}$$

####    2a) ¿Es $H$ normal en $G$? Encuentra $G/H$.

Para probar que nuestro subgrupo es normal, vamos a aprovechar la siguiente definición:

$$H \text{  es Normal  } \quad \Longleftrightarrow \quad \forall_{h \in H}\,\, \forall_{g \in G} \,\,\, g  h g^{-1} \, \in H$$

Adicionalmente, vamos a aprovechar que la operación suma es conmutativa y la definición de inverso por notación de suma.

$$ a + b = b + a \quad \forall_{a,b \in \mathbb{Z}}$$

$$\forall_{a \in G} \, \exists_{-a \in G} \quad a + (-a) = (-a) + a = 0$$

Suponemos la existencia de un par de elementos independientes que pertenecen a $G$ y a $H$ respectivamente; y a partir de ellos, vamos a denotar si la relación expuesta en la definición de subgrupo normal se cumple.

Dado $h \in H$ y $g \in G$:

$$
\begin{array}{rclcr}
   g \, h \,  g^{-1} & = & g + h + (-g) & &\text{Notación de }+ \\
                     & = & g + (-g) + h & &\text{Propiedad Conmutativa } \\
                     & = & 0 + h & &\text{Propiedad inverso} \\
                     & = & h \in H & &\text{Propiedad identidad} \\
\end{array}
$$

Ya sabiendo que el subgrupo $H$ es normal, vamos a hallar todas sus posibles clases laterales para determinar $G/H$

$$H + 0 = H = \{...,-20, -15 , -10 -5, 0, 5 ,10 ,15 , 20 , ...\}$$
$$H + 1 = H_{1} = \{...,-19, -14 , -9 -4, 1, 6 ,11 ,16 , 21 , ...\}$$
$$H + 2 = H_{2} = \{...,-18, -13 , -8 -3, 2, 7 ,12 ,17 , 22 , ...\}$$
$$H + 3 = H_{3} = \{...,-17, -12 , -7 -2, 3, 8 ,13 ,18 , 23 , ...\}$$
$$H + 4 = H_{4} = \{...,-16, -11 , -6 -1, 4, 9 ,14 ,19 , 24 , ...\}$$

Hemos encontrado todas las posibles clases laterales disyuntas.

$$G = H \cup H_{1} \cup H_{2} \cup H_{3} \cup H_{4}$$

$$G/H = \{ H \, ,\, H_{1} \, ,\, H_{2} \, ,\, H_{3} \, ,\, H_{4}\}$$

Podemos concluir:

$$\therefore H \text{  es Normal  } \quad \wedge \quad G/H = \{ H \, ,\, H_{1} \, ,\, H_{2} \, ,\, H_{3} \, ,\, H_{4}\}$$
   
####    2b) ¿Es $G/H$ cíclico?

Para saber si el grupo $G/H$ es cíclico, debemos encontrar un generador o demostrar que no dispone de uno.

Vamos a tomar de candidato $H_{1}$ y vamos a tener en cuenta que al operar los elementos, aplicaremos la operación $\oplus$ (suma módulo 5) para no tener inconvenientes al cambiar el representante de la clase.

$$
\begin{array}{rcl}
   \left(H_{1}\right)^{1} &=& H_{1} \\
   \left(H_{1}\right)^{2} &=& H_{1 \oplus 1} = H_{2} \\
   \left(H_{1}\right)^{3} &=& H_{2 \oplus 1} = H_{3} \\
   \left(H_{1}\right)^{4} &=& H_{3 \oplus 1} = H_{4} \\
   \left(H_{1}\right)^{5} &=& H_{4 \oplus 1} = H \\
\end{array}
$$

Podemos denotar que a partir de $H_{1}$ podemos generar todo el grupo. Es decir:

$$ G/H = \langle H_{1} \rangle$$

Podemos concluir:

$$\therefore G/H \text{ es cíclico}$$
