##### Códigos de Apoyo

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def map_plot():
    domain = [i for i in range(8)]
    codomain = [i for i in range(4)]

    mapping = {i: i%4 for i in domain}

    fig, ax = plt.subplots(figsize=(8, 4))

    domain_oval = patches.Ellipse((1, 1.25), 2, 3.5, edgecolor='blue', facecolor='none')
    codomain_oval = patches.Ellipse((4, 1.25), 2, 3, edgecolor='red', facecolor='none')
    ax.text(1, 3.25, '$(\\mathbb{Z}_8,\\oplus)$', fontsize=14, ha='center',color='blue')
    ax.text(4, 3, '$(\\mathbb{Z}_4,\\oplus)$', fontsize=14, ha='center',color='red')
    ax.add_patch(domain_oval)
    ax.add_patch(codomain_oval)

    domain_coords = [[1, i*0.35] for i in range(len(domain)) ]
    codomain_coords = [[4, 0.5+i*0.5] for i in range(len(codomain))]

    for i, element in enumerate(domain):
        ax.text(domain_coords[i][0], domain_coords[i][1], str(element), fontsize=12, ha='center', va='center')

    for i, element in enumerate(codomain):
        ax.text(codomain_coords[i][0], codomain_coords[i][1], str(element), fontsize=12, ha='center', va='center')

    for i, (d, c) in enumerate(mapping.items()):
        start = domain_coords[domain.index(d)]
        end = codomain_coords[codomain.index(c)]
        start[0]+=0.2
        end[0]-=0.1
        ax.annotate("",
                    xy=end, xycoords='data',
                    xytext=start, textcoords='data',
                    arrowprops=dict(arrowstyle="->", lw=1.5))

    ax.set_xlim(-1, 6)
    ax.set_ylim(-1, 4)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_aspect('equal')
    ax.set_title('Mapeo de  $\\varphi: (\\mathbb{Z}_8,\\oplus) \\, \\to \\, (\\mathbb{Z}_4,\\oplus)$')
    plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def kernel_plot():
    domain = [i for i in range(8)]
    codomain = [i for i in range(4)]

    mapping = {i: i%4 for i in domain}

    fig, ax = plt.subplots(figsize=(8, 4))

    domain_oval = patches.Ellipse((1, 1.25), 2, 3.5, edgecolor='blue', facecolor='none')
    codomain_oval = patches.Ellipse((4, 1.25), 2, 3, edgecolor='red', facecolor='none')
    ax.text(1, 3.25, '$(\\mathbb{Z}_8,\\oplus)$', fontsize=14, ha='center',color='blue')
    ax.text(4, 3, '$(\\mathbb{Z}_4,\\oplus)$', fontsize=14, ha='center',color='red')
    ax.add_patch(domain_oval)
    ax.add_patch(codomain_oval)

    domain_coords = [[1, i*0.35] for i in range(len(domain)) ]
    codomain_coords = [[4, 0.5+i*0.5] for i in range(len(codomain))]

    for i, element in enumerate(domain):
        if mapping[element]==0:
            ax.text(domain_coords[i][0], domain_coords[i][1], str(element), fontsize=12, ha='center', va='center',color='green')
        else:
            ax.text(domain_coords[i][0], domain_coords[i][1], str(element), fontsize=12, ha='center', va='center')

    for i, element in enumerate(codomain):
        if element==0:
            ax.text(codomain_coords[i][0], codomain_coords[i][1], str(element), fontsize=12, ha='center', va='center',color='green')
        else:
            ax.text(codomain_coords[i][0], codomain_coords[i][1], str(element), fontsize=12, ha='center', va='center')

    for i, (d, c) in enumerate(mapping.items()):
        start = domain_coords[domain.index(d)]
        end = codomain_coords[codomain.index(c)]
        start[0]+=0.2
        end[0]-=0.1
        if c==0:
            ax.annotate("",
                        xy=end, xycoords='data',
                        xytext=start, textcoords='data',
                        arrowprops=dict(arrowstyle="->", lw=1.5,color='green'))
        else:
            ax.annotate("",
                        xy=end, xycoords='data',
                        xytext=start, textcoords='data',
                        arrowprops=dict(arrowstyle="->", lw=1.5))

    ax.set_xlim(-1, 6)
    ax.set_ylim(-1, 4)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_aspect('equal')
    ax.set_title('Kernel de  $\\varphi: (\\mathbb{Z}_8,\\oplus) \\, \\to \\, (\\mathbb{Z}_4,\\oplus)$')
    ax.scatter([], [], color='green', label='$ker(\\varphi)$')
    ax.legend()
    plt.show()
```

### Dado $\varphi:(\mathbb{Z}_8,\oplus)\rightarrow (\mathbb{Z}_4,\oplus)$ definido como $\varphi(x)= x \, \text{mod} \,4$

Antes de ir a responder las preguntas, vamos a desglosar el homomorfismo descrito de tal forma que podamos identificar como están mapeandose los elementos de un grupo a otro.

En primer lugar, visualizaremos cuantos y cuales elementos disponen ambos grupos:

$$(\mathbb{Z}_8,\oplus) = \{0,1,2,3,4,5,6,7\}$$

$$(\mathbb{Z}_4,\oplus) = \{0,1,2,3\}$$

Dado que ambos grupos son finitos y particularmente pequeños, vamos a visualizar en un diagrama de Venn las relaciones entre los dos grupos:

```python
map_plot()
```

Dado el mapeo, podemos darnos cuenta que el homomorfismo cumple con la condición de ser sobre aunque no es uno a uno; por lo que es un epimorfismo pero no un isomorfismo

#### 1. Encuentre $ker(\varphi)$

Para encontrar el kernel del homomorfismo, debemos identificar que elementos del grupo de salida mapean a la identidad del conjunto de llegada, es decir:

$$ker(\varphi) = \{x \in \mathbb{Z}_8 : \varphi(x)=0\}$$

Para fines practivos, vamos a identificar este kernel de manera gráfica:

```python
kernel_plot()
```

A partir del mapeo, podemos identificar que el kernel de $\varphi$ está compuesto de dos elementos, los cuales son 4 y 0.

$$\therefore ker(\varphi) = \{0,4\}$$

#### 2. ¿A qué grupo familiar es $(\mathbb{Z}_8,\oplus)/ker(\varphi)$ isomórfico?

El primer paso será determinar cual es el grupo $(\mathbb{Z}_8,\oplus)/ker(\varphi)$ antes de decidir o mencionar a que otro grupo es isomórfico.

Dado que ya conocemos el que el kernel de $\varphi$ es un subgrupo de $\mathbb{Z}_8$, determinaremos sus clases laterales.
Definiendo $H = ker(\varphi)$:

$$H_{0} = \{0,4\}$$
$$H_{1} = \{1,5\}$$
$$H_{2} = \{2,6\}$$
$$H_{3} = \{3,7\}$$

Dado esto, podemos definir que el grupo $(\mathbb{Z}_8,\oplus)/ker(\varphi)$ esta dado por:

$$(\mathbb{Z}_8,\oplus)/ker(\varphi) = \{H , H_1, H_2, H_3 \} \quad \text{donde}\quad H = ker(\varphi)$$

Para determinar con que grupo es isomorfico el grupo encontrado, vamos a realizar la tabla de mulriplicación de todos los elementos del grupo:

\begin{array}{|r||c|c|c|c|}
    \hline
      & H & H_{1} & H_{2} & H_{3}\\
    \hline
    \hline
    H & H & H_1 & H_2 & H_3\\
    \hline
    H_1 & H_1 & H_2 & H_3 & H \\
    \hline
    H_2 & H_2 & H_3 & H & H_1 \\
    \hline
    H_3 & H_3 & H & H_1 & H_2 \\
    \hline
\end{array}

Adicionalmente, y apoyandonos en la misma tabla, vamos a hallar el orden de los elementos del grupo, denotando que la identidad del grupo es $H$:

\begin{array}{lcr}
  H = H & \Longrightarrow & o(H) = 1 \\
  H_1 \, H_1 = H_2 \, H_1 = H_3 \, H_1 = H & \Longrightarrow & o(H_1)=4 \\
  H_2 \, H_2 = H & \Longrightarrow & o(H_2)=2 \\
  H_3 \, H_3 = H_2 \, H_3 = H_1 \, H_3 = H & \Longrightarrow & o(H_3)=4 \\
\end{array}

A partir de los teoremas asociados a subgrupos normales e isomorfismos, podemos concluir que el grupo $(\mathbb{Z}_4,\oplus)$ mapeado por $\varphi$ es isomórfico con el grupo $(\mathbb{Z}_8,\oplus)/ker(\varphi)$. Dado que esta relación esta contemplada por un teorema, no nos centraremos en explicarla a mucho detalle. Sin embargo, aprovecharemos una convención gráfica que nos permita visualizarla sin mucho inconveniente. Para la convención gráfica, aprovecharemos el isomorfismo descrito por:

$$f(H_{x}) = \varphi(x)$$

```python
iso_plot()
```

$$\therefore \quad (\mathbb{Z}_8,\oplus)/ker(\varphi) \cong (\mathbb{Z}_4,\oplus)$$
