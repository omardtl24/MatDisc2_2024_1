# Definción del Problema

Los diodos son componentes electrónicos fundamentales que permiten el flujo de corriente en una sola dirección, actuando como válvulas unidireccionales para la corriente eléctrica.

Su importancia en el mundo moderno es inmensa, ya que se utilizan en una amplia variedad de aplicaciones tecnológicas. Desde la rectificación de corriente alterna a corriente continua en fuentes de alimentación, hasta su uso en diodos emisores de luz (LED) en pantallas y sistemas de iluminación, los diodos son esenciales para el funcionamiento eficiente de numerosos dispositivos electrónicos. Además, juegan un papel crucial en la protección de circuitos, la regulación de voltaje y en la conversión de energía en sistemas de energía renovable. La capacidad de los diodos para controlar el flujo de corriente y su integración en diversos sistemas hacen de ellos componentes indispensables en la electrónica moderna.

En este contexto, el problema que se va abordar está relacionado con el comportamiento físico-matemático que está detrás de estos distopisitivos semiconductores.

Lo primero que haremos es definir que es un diodo. Un diodo es un componente tridimensional que está compuesto por dos elementos principales, el ánodo y el cátodo. Este componente se caracteriza por permitir el paso de un flujo de electrones en una dirección fija. El siguiente gráfico permite ver la distribución de estos elementos en un plano tridimensional.

![Esquema](./Esquema.png)


En este caso, los electrones fluyen únicamente desde el cátodo hasta el ánodo. Para que este fenómeno suceda, deben suceder diversos factores físicos que por el alcance de este trabajo no se van a especificar. De todas formas, es importante resaltar que este comportamiento ha sido analizado por distintos fisicos y matematicos a lo largo del último siglo.

Gracias a los aportes Langmuire y Compton, con el llamado sistema de ecuaciones diferenciales de Vlasov-Maxwell:

$$
\begin{array}{c}
  \dfrac{d^2\varphi}{dx^2} = j_x \dfrac{1+\varphi}{\sqrt{(1+\varphi)^2 -1 - a^2}}\\
  \dfrac{d^2a}{dx^2} = j_x \dfrac{a}{\sqrt{(1+\varphi)^2 -1 - a^2}}
\end{array}
$$

con las sisguientes condiciones iniciales y de frontera:

$$
\begin{array}{c}
  a(0)=0 & a(1)=a_L & a'(0)=\alpha \\
  \varphi(0)=0 & \varphi(1)=a_L & \varphi'(0)=\beta
\end{array}
$$

donde:

* $x$: Coordenada en el que se encuentra el electrón, representa la distancia relativa entre el ánodo ($x=1$) y el cátodo ($x=0$)
* $a(x)$: Función que denota el potencial magnético de acuerdo al punto $x$
* $\varphi(x)$: Función que describe el cambio en el campo eléctrico de acuerdo al punto $x$
* $j_x$: Parámetro que define la densidad de la corriente
* $\beta$: Parámetro que carcateriza el campo magnético cerca al cátodo
* $\alpha$: Parámetro que carcateriza el potencial magnético cerca al cátodo

A partir de este sistema, se pueden realizar una serie de reducciones en base al potencial efectivo del diodo $\theta$, el cual está definido por:

$$\theta = (1+\varphi)^2 -1 - a^2$$

Obteniendo así la siguiente ecuación diferencial con sus condiciones de frontera:

$$\left(\dfrac{d \theta}{dx}\right)^2 = k \theta + 8j_x \theta ^{\frac{3}{2}} + 8j_x \theta ^{\frac{1}{2}} + 4\beta^2$$

$$\theta(0) = 0 \quad \quad \theta(1)=(1+\varphi_L)^2 -1 - a_L^2 = \theta_L $$



Adicionalmente, se va a trabajar un fenómeno particular denominado aislamiento magnético. Este fenómeno consiste en que, a pesar de que exista un flujo correcto de electrones que salen del cátodo, los electrones nunca llegan al ánodo, sino que regresan al cátodo debido a factores asociados al campo magnético.

El experimento del tubo de rayos catódicos es una forma sencilla y práctica de visualizar el accionar de este fenómeno. Al aplicarle un campo magnetico a un flujo de elecrones, es posible cambiar su dirección.

![GIF Description](./Newton_Exp.gif)


La forma de asociar este fenomeno al planteamiento, es determinar en que momento de la trayectoria del electron se presentará un decrecimiento en su potencial efectivo $\theta$. Es decir, en que momento se presenta la situación:

$$\dfrac{d\theta}{dx} = 0$$

Utilizando la expresión reducida, podemos llevarlo a que nuestro problema se limita a dar solución a la ecuación no lineal:

$$k \theta + 8j_x \theta ^{\frac{3}{2}} + 8j_x \theta ^{\frac{1}{2}} + 4\beta^2 = 0$$

Dada la complejidad de la expresión se aplicó la sustitución $\theta = u^2$ y se redujeron los parametros de la expresión a los siguientes:

$$\hat{k} = \dfrac{k}{8j_x} \quad \quad \hat{\beta} = \dfrac{\beta^2}{2j_x}$$

obteniento finalmente la siguiente ecuación de tercer grado:

$$u^3 + \hat{k}u^2+u+\beta = 0$$

# Solución Analítica

Ya definido que toda ecuación cúbica tiene 3 soluciones complejas (Con multiplicidad en algunos casos), nos interesa llevar este hecho al planteamiento de aislamento magnetico del diodo.

Para ello, es importante definir los casos en los que podemos tener tres soluciones diferentes, dos soluciones diferentes (Una solución con multiplicidad dos) o una única solución (Multiplicidad 3). Esto lo define el discriminante, el cual está dado por:

$$\Delta_u = 18\hat{k}\hat{\beta} + \hat{k}^2 - 4 - 4\hat{k}^3\hat{\beta} - 27 \hat{\beta}^2$$

A partir de este, los diferentes casos son:


##$\Delta_u < 0$

Para este primer caso, se van a a obtener tres soluciones, una solución real y dos soluciones complejas conjugadas. Para ello, el proceso a aplicar es obtener la forma reducida de la expresión para posteriormente aplicar la regla de Cardano.

El primer paso es obtener la forma reducida de la expresión cúbica usando la transformación de Tschirnhaus definida por $u = y-\frac{\hat{k}}{3}$

$$\left(y - \dfrac{\hat{k}}{3}\right)^3 + \hat{k} \, \left(y - \dfrac{\hat{k}}{3}\right)^2 +  \left(y - \dfrac{\hat{k}}{3}\right)  + \hat{\beta} = 0 $$

$$y^3 - \hat{k}y^2 + \dfrac{\hat{k}^2y}{3} - \dfrac{\hat{k}^3}{27} + \hat{k} y^2 - \dfrac{2\hat{k}^2y}{3} + \dfrac{\hat{k}^3}{9} + y - \dfrac{\hat{k}}{3}  + \hat{\beta} = 0 $$

$$y^3 + \left( \dfrac{\hat{k}^2}{3}- \dfrac{2\hat{k}^2}{3} + 1\right) y  + \left(- \dfrac{\hat{k}^3}{27}   + \dfrac{\hat{k}^3}{9} - \dfrac{\hat{k}}{3}  + \hat{\beta}\right) = 0 $$

$$y^3 + \left( \dfrac{3 - \hat{k}^2}{3}\right) y  + \left(\dfrac{2\hat{k}^3 - 9\hat{k} + 27\hat{\beta}}{27} \right) = 0 $$

Adicionalmente, aplicamos la regla de cardano definida por la expresión:

$$(m+n)^3 - 3mn(m+n) - (m^3 + n^3) = 0$$

Por consiguiente, en nuestro contexto obtenemos que:

$$y = m+n \,\,\,\, , \,\,\,\, -\dfrac{3 - \hat{k}^2}{9} = mn \,\,\,\, , \,\,\,\,-\dfrac{2\hat{k}^3 - 9\hat{k} + 27\hat{\beta}}{27} = m^3 + n^3$$

A partir de una serie de operaciones algebraicas, obtenemos la siguiente expresión:

$$729n^6 +\left( 54\hat{k}^3 - 243\hat{k} + 729\hat{\beta} \right)n^3 - \left(3 - \hat{k}^2\right)^3 = 0$$

A esta expresión le aplicamos la formula de la ecuación cúbica para obtener un resultado para $n^3$:

$$n^3 = \dfrac{1}{1458}\left(-\left( 54\hat{k}^3 - 243\hat{k} + 729\hat{\beta} \right) \pm \sqrt{\left( 54\hat{k}^3 - 243\hat{k} + 729\hat{\beta} \right)^2 + 2916 \left(3 - \hat{k}^2\right)^3 }\right)$$

Para esta expresión, nos interesa saber si su discriminante es positivo (Es decir que todo resultado es real). Es decir, que queremos verificar que $\left( 54\hat{k}^3 - 243\hat{k} + 729\hat{\beta} \right)^2 + 2916 \left(3 - \hat{k}^2\right)^3 > 0$. Podemos desglosar la expresión de la siguiente forma:

$$\left(54\hat{k}^3 - 243\hat{k} + 729\hat{\beta} \right)^2 + 2916 \left(3 - \hat{k}^2\right)^3 = -19638\left( 18\hat{k}\hat{\beta} + \hat{k}^2 - 4 - 4\hat{k}^3\hat{\beta} - 27 \hat{\beta}^2 \right) = -19638 \Delta_u$$

Y dado que $\Delta_u < 0$ podemos concluir que $-19638\Delta_u > 0$

Adiocionalmente, sustituimos nuestra expresión en la ecuación de Cardano para obtener el valor de $m^3$

$$-\dfrac{2\hat{k}^3 - 9\hat{k} + 27\hat{\beta}}{27} = m^3 + n^3$$

$$-\dfrac{2\hat{k}^3 - 9\hat{k} + 27\hat{\beta}}{27} = m^3 + \dfrac{1}{1458}\left(-\left( 54\hat{k}^3 - 243\hat{k} + 729\hat{\beta} \right) \pm \sqrt{\left( 54\hat{k}^3 - 243\hat{k} + 729\hat{\beta} \right)^2 + 2916 \left(3 - \hat{k}^2\right)^3 }\right)$$
$$ $$
$$-\dfrac{2\hat{k}^3 - 9\hat{k} + 27\hat{\beta}}{27} - \dfrac{1}{1458}\left(-\left( 54\hat{k}^3 - 243\hat{k} + 729\hat{\beta} \right) \pm \sqrt{\left( 54\hat{k}^3 - 243\hat{k} + 729\hat{\beta} \right)^2 + 2916 \left(3 - \hat{k}^2\right)^3 }\right)= m^3$$
$$ $$
$$ \dfrac{1}{1458}\left(-\left( 54\hat{k}^3 - 243\hat{k} + 729\hat{\beta} \right) \mp \sqrt{\left( 54\hat{k}^3 - 243\hat{k} + 729\hat{\beta} \right)^2 + 2916 \left(3 - \hat{k}^2\right)^3 }\right)= m^3$$

A partir de esto, esto, podemos obtener los valores de $m$ y $n$ respectivamente:

$$m = \dfrac{\sqrt[3]{4}}{18}\sqrt[3]{ -\left( 54\hat{k}^3 - 243\hat{k} + 729\hat{\beta} \right) - \sqrt{\left( 54\hat{k}^3 - 243\hat{k} + 729\hat{\beta} \right)^2 + 2916 \left(3 - \hat{k}^2\right)^3 }}$$

Podemos definir la raíz real y las raíces complejas conjugadas de la expresión en terminos de $y$ utilizando las raices cúbicas de la unidad.

Finalmente, aplicamos al transformada inversa de Tschirnhaus, obteniendo así las soluciones en términos de $u$:

$$u_i  = y_i - \dfrac{\hat{k}}{3}    \,\,\,\,\,\,\,\,\, \forall i \in \{1,2,3\}$$

$$u_2 = \left[ - \dfrac{\hat{k}}{3}  - \dfrac{\sqrt[3]{4}}{36} \left( \sqrt[3]{A_1 + A_2} + \sqrt[3]{A_1 - A_2} \right) \right] + \left[ \dfrac{\sqrt{3}\sqrt[3]{4}}{36} \left( \sqrt[3]{A_1 + A_2} - \sqrt[3]{A_1 - A_2} \right) \right]i$$

##$\Delta_u = 0$

En este caso, tendremos únicamente soluciones reales que cumplen con la propiedad de tener multiplicidad mayor a uno. Es decir, que nuestras soluciones van a ser de la forma:

\begin{array}{rcl}
  (u-a)(u-b)^2 = 0 & & \text{Solución de multiplicidad } 2 \\
  (u-b)^3 = 0 & & \text{Solución de multiplicidad } 3
\end{array}

Pos consiguiente, analizaremos estos dos casos por aparte:

### Caso 1: Multiplicidad 2

Para este primer caso, vamos a tomar la forma reducida de nuestra expresión obtenida de la transformación de Tschirnhaus.

$$y^3 + \left( \dfrac{3 - \hat{k}^2}{3}\right) y  + \left(\dfrac{2\hat{k}^3 - 9\hat{k} + 27\hat{\beta}}{27} \right) = 0 $$

Vamos a suponer que esta expresión viene de dos raices y vamos a ampliar la expresion:

\begin{array}{rcl}
  y^3 + \left( \dfrac{3 - \hat{k}^2}{3}\right) y  + \left(\dfrac{2\hat{k}^3 - 9\hat{k} + 27\hat{\beta}}{27} \right) &=& (y-y_1)(y-y_2)^2 \\
  & \vdots & \\
  y^3 + \left( \dfrac{3 - \hat{k}^2}{3}\right) y  + \left(\dfrac{2\hat{k}^3 - 9\hat{k} + 27\hat{\beta}}{27} \right) &=& y^3 + (-2y_2 -y_1)y^2 + (y_2^2 + 2y_1y_2)y + (-y_2^2y_1)
\end{array}

A partir de esta igualdad obtenemos el siguiente sistema de ecuaciones no lineales:

$$-2y_2 -y_1 = 0$$
$$ y_2^2 + 2y_1y_2 = \dfrac{3 - \hat{k}^2}{3}$$
$$-y_2^2y_1 = \dfrac{2\hat{k}^3 - 9\hat{k} + 27\hat{\beta}}{27}$$

Al ser un sistema simple, lo resolvemos con reemplazos y manipulación algebraica, obteniendo los siguientes resultados para $y_1$ y $y_2$.

$$y_1 =  \dfrac{2\hat{k}^3 - 9\hat{k} + 27\hat{\beta}}{9 - 3\hat{k}^2}$$
$$y_2  =  \dfrac{2\hat{k}^3 - 9\hat{k} + 27\hat{\beta}}{6\hat{k}^2 - 18}$$

Finalmente aplicamos la transformada inversa de Tschirnhaus para obtener los resultados en términos de $u$

$$u_1 = y_1 - \frac{\hat{k}}{3} =\dfrac{2\hat{k}^3 - 9\hat{k} + 27\hat{\beta}}{9 - 3\hat{k}^2} - \frac{\hat{k}}{3}$$
$$ $$
$$u_2 = y_2 - \frac{\hat{k}}{3} =\dfrac{2\hat{k}^3 - 9\hat{k} + 27\hat{\beta}}{6\hat{k}^2 - 18} - \frac{\hat{k}}{3}$$

### Caso 2: Multiplicidad 3

Para el caso de multiplicidad 3, vamos a partir de la expresión cúbica inicial y la estructura de la solución.

\begin{array}{rcl}
 u^3 + \hat{k} \, u^2 +  u  + \hat{\beta} &=& (u - u_1)^3 \\
  & \vdots & \\
  u^3 + \hat{k} \, u^2 +  u  + \hat{\beta} &=& u^3 - 3u^2u_1 + 3uu_1^2 + u_1^3
\end{array}

A partir de esta igualdad, obtenemos el siguiente sistema de ecuaciones no lineales:

$$3u_1 = - \hat{k}$$
$$3u_1^2 = 1$$
$$u_1^3 = - \hat{\beta}$$

A partir de este, podemos concluir que solo existen dos soluciones para nuestro problema:

$$\hat{\beta} =   \dfrac{\sqrt{3}}{9} \,\,\,  \land \,\,\, \hat{k} =  \sqrt{3} \,\,\, \longrightarrow \,\,\, u_1 = - \dfrac{ \hat{k}}{3} = - \dfrac{\sqrt{3}}{3}$$
$$ $$
$$\hat{\beta} =  - \dfrac{\sqrt{3}}{9} \,\,\,  \land \,\,\, \hat{k} = - \sqrt{3} \,\,\, \longrightarrow \,\,\, u_1 = - \dfrac{ \hat{k}}{3} = \dfrac{\sqrt{3}}{3}$$



##$\Delta_u > 0$

Para este caso, vamos tener tres soluciones reales. Por la teoría de Galouis, sabemos que no hay forma de escribir esto sin usar los complejos (No se puede escribir con sumas prosuctos y radicales sin el componente complejo). Dado esto, usaremos las soluciones trigonométricas de Viéte para expresar las souciones como números reales sin complejos.

El primer paso es aplicar la transformación de Tschirnhaus una vez más:

$$y^3 + \left( \dfrac{3 - \hat{k}^2}{3}\right) y  + \left(\dfrac{2\hat{k}^3 - 9\hat{k} + 27\hat{\beta}}{27} \right) = 0 $$

A partir de este, aplicamos la formula descrita por:

$$z^3 + pz + q = 0 \,\,\,\,\,\, \longrightarrow \,\,\,\,\,\, z_{t+1} = 2\sqrt{-\dfrac{p}{3}} \text{cos}\left( \dfrac{1}{3} \text{arccos}\left( \dfrac{3q}{2p} \sqrt{\dfrac{-3}{p}}\right) + \dfrac{2t\pi}{3}\right)\,\,\,\, \forall t \in \{0,1,2\}$$

Lo que en nuestro caso genera los siguientes valores:

$$\sqrt{-\dfrac{p}{3}} = \sqrt{-\dfrac{\frac{3 - \hat{k}^2}{3}}{3}} = \sqrt{\dfrac{\hat{k}^2-3}{9}}$$

$$\sqrt{-\dfrac{3}{p}} = \sqrt{-\dfrac{3}{\frac{3 - \hat{k}^2}{3}}} = \sqrt{\dfrac{9}{\hat{k}^2-3}}$$

$$\dfrac{3q}{2p} = \dfrac{3\left( \frac{\hat{k}^3}{27}   + \dfrac{\hat{k}^3}{9} - \dfrac{\hat{k}}{3}  + \hat{\beta} \right)}{2\left( \frac{3 - \hat{k}^2}{3} \right)} = \dfrac{4\hat{k}^3-9\hat{k}+27\hat{\beta}}{18 - 6\hat{k}^2}$$

Dando las siguientes soluciones:

$$y_{1} = 2\sqrt{\dfrac{\hat{k}^2-3}{9}} \text{cos}\left( \dfrac{1}{3} \text{arccos}\left( \dfrac{4\hat{k}^3-9\hat{k}+27\hat{\beta}}{18 - 6\hat{k}^2} \sqrt{\dfrac{9}{\hat{k}^2-3}}\right)\right)$$

$$y_{2} = 2\sqrt{\dfrac{\hat{k}^2-3}{9}} \text{cos}\left( \dfrac{1}{3} \text{arccos}\left( \dfrac{4\hat{k}^3-9\hat{k}+27\hat{\beta}}{18 - 6\hat{k}^2} \sqrt{\dfrac{9}{\hat{k}^2-3}}\right) + \dfrac{2\pi}{3}\right)$$

$$y_{3} = 2\sqrt{\dfrac{\hat{k}^2-3}{9}} \text{cos}\left( \dfrac{1}{3} \text{arccos}\left( \dfrac{4\hat{k}^3-9\hat{k}+27\hat{\beta}}{18 - 6\hat{k}^2} \sqrt{\dfrac{9}{\hat{k}^2-3}}\right) + \dfrac{4\pi}{3}\right)$$

Al aplicar la transformada inversa de Tschirnhaus obtenemos los siguientes resultados:

$$u_{1} =\dfrac{2}{3}\sqrt{\hat{k}^2-3}\,\,\,\text{cos}\left( \dfrac{1}{3} \text{arccos}\left( \dfrac{4\hat{k}^3-9\hat{k}+27\hat{\beta}}{18 - 6\hat{k}^2} \sqrt{\dfrac{9}{\hat{k}^2-3}}\right)\right) - \dfrac{\hat{k}}{3}$$

$$u_{2} = \dfrac{2}{3}\sqrt{\hat{k}^2-3}\,\,\, \text{cos}\left( \dfrac{1}{3} \text{arccos}\left( \dfrac{4\hat{k}^3-9\hat{k}+27\hat{\beta}}{18 - 6\hat{k}^2} \sqrt{\dfrac{9}{\hat{k}^2-3}}\right) + \dfrac{2\pi}{3}\right) - \dfrac{\hat{k}}{3}$$

$$u_{3} = \dfrac{2}{3}\sqrt{\hat{k}^2-3}\,\,\, \text{cos}\left( \dfrac{1}{3} \text{arccos}\left( \dfrac{4\hat{k}^3-9\hat{k}+27\hat{\beta}}{18 - 6\hat{k}^2} \sqrt{\dfrac{9}{\hat{k}^2-3}}\right) + \dfrac{4\pi}{3}\right) - \dfrac{\hat{k}}{3}$$














# Implementación y Visualización

Para poder realizar una correcta implementación y visualización,se debe definir la función que describe los resultados analiticos que dan solución a la ecuación cúbica

```python
import numpy as np

cubic_poly = lambda k,b,u : u**3 + k*u**2 + u + b

def u_values(k, b):
    delta = 18*k*b + k**2 - 4 - 4*k**3*b - 27*b**2

    if delta == 0:
        if (k == np.sqrt(3) and b == np.sqrt(3)/9) or (k == -np.sqrt(3) and b == -np.sqrt(3)/9):
            return np.array([-k/3, -k/3, -k/3])
        else:
            arg = 2*k**3 - 9*k + 27*b
            u1 = (arg / (9 - 3*k**2)) - k/3
            u2 = (arg / (6*k**2 - 18)) - k/3
            return np.array([u1, u2, u2])
    elif delta > 0:
        coef = (2/3) * np.sqrt(k**2 - 3)
        arg = ((2*k**3 - 9*k + 27*b) / (18 - 6*k**2)) * np.sqrt(9 / (k**2 - 3))
        u1 = coef * np.cos((1/3) * np.arccos(arg)) - k/3
        u2 = coef * np.cos((1/3) * np.arccos(arg) + (2*np.pi/3)) - k/3
        u3 = coef * np.cos((1/3) * np.arccos(arg) + (4*np.pi/3)) - k/3
        return np.array([u1, u2, u3])
    else:
        A1 = -(54*k**3 - 243*k + 729*b)
        A2 = np.sqrt(-19683*delta)
        C = np.cbrt
        u1 = -k/3 + (C(4)/18) * (C(A1 + A2) + C(A1 - A2))
        u2_real = -k/3 - (C(4)/36) * (C(A1 + A2) + C(A1 - A2))
        u2_imag = (np.sqrt(3) * C(4) / 36) * (C(A1 + A2) - C(A1 - A2))
        u2 = u2_real + 1j * u2_imag
        u3 = u2_real - 1j * u2_imag
        return np.array([u1, u2, u3])
```

Verificamos que la solución se mantenga en un valor cercano a cero, se mantendrá una tolerancia de $10^{-10}$

```python
import time

start = time.time()
errhigh = []
cases = 0
for k in range(-50,50,1):
    k = k/10
    for b in range(-50,50,1):
        b = b/10
        u = u_values(k,b)
        for r in u:
            cases += 1
            t = abs(cubic_poly(k,b,r))<1e-10
            if not t:
                errhigh.append((k,b))
end = time.time()
print(f'Time used: {(end-start):.4f} seconds')
print(f'After {cases} cases evaluated, {len(errhigh)} cases had a wrong value')
if len(errhigh)>0:
  print(f'Wrong cases: {errhigh}')
```

Dado que verificamos que nuestras soluciones están funcionando correctamente, diseñaremos una rutina que nos permita obtener una grilla de soluciones que guardaremos en un dataFrame de para posteriormente graficarlos



```python
import numpy as np
import pandas as pd

def get_fixed_k(k=1, minb=-1,maxb=1,num_points = 100):

  bparams = np.linspace(minb,maxb,num_points)
  expanded_results = {'b': [], 'u_real': [], 'u_complex':[]}

  for b in bparams:
    u_vals = u_values(k, b)
    for u in u_vals:
      expanded_results['b'].append(b)
      expanded_results['u_real'].append(np.real(u))
      expanded_results['u_complex'].append(np.imag(u))

  final_df =  pd.DataFrame(expanded_results)
  final_df['k'] = k
  return final_df

def get_fixed_b(b=1, minb=-1,maxb=1,num_points = 100):

  kparams = np.linspace(minb,maxb,num_points)
  expanded_results = {'k': [], 'u_real': [], 'u_complex':[]}

  for k in kparams:
    u_vals = u_values(k, b)
    for u in u_vals:
      expanded_results['k'].append(k)
      expanded_results['u_real'].append(np.real(u))
      expanded_results['u_complex'].append(np.imag(u))

  final_df =  pd.DataFrame(expanded_results)
  final_df['b'] = b
  return final_df

def get_grid(mink = -1, maxk = 1, minb = -1, maxb = 1, num_points = 100):

  kparams = np.linspace(mink,maxk,num_points)
  bparams = np.linspace(minb,maxb,num_points)
  K,B = np.meshgrid(kparams, bparams)

  k_flat = K.flatten()
  b_flat = B.flatten()

  expanded_results = {'k': [], 'b': [], 'u_real': [], 'u_complex':[]}

  for k,b in zip(k_flat, b_flat):
    u_vals = u_values(k, b)
    for u in u_vals:
        expanded_results['k'].append(k)
        expanded_results['b'].append(b)
        expanded_results['u_real'].append(np.real(u))
        expanded_results['u_complex'].append(np.imag(u))

  final_df =  pd.DataFrame(expanded_results)
  return final_df
```

## Primeras Gráficas

Diseñamos un par de rutinas que nos permitan visualizar la parte real e imaginaria de cada sección tomando como punto de partida dejar un parametro fijo (sea $\hat{k}$ o $\hat{\beta}$)

```python
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

def plot_fixed_k(df,k = 0,klabel = None, mark_point = None):
  assert k in df['k'].values
  try:
    if klabel is None:
      klabel = str(k)
    filtered_df = df[df['k']==k]
    df_real = filtered_df[filtered_df['u_complex']==0]
    df_com = filtered_df[filtered_df['u_complex']!=0]

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

    ax1.scatter(df_real['b'], df_real['u_real'], color='red', s=1 , label='Real solutions')
    ax1.scatter(df_com['b'], df_com['u_real'], color='blue', s=1 , label='Real part (Complex solution)')
    if mark_point is not None:
      ax1.scatter(mark_point[0], mark_point[1], color='green', s=10 , label=mark_point[3])
    ax1.set_title('Real part plot')
    ax1.set_xlabel('$\\hat{\\beta}$')
    ax1.set_ylabel('Re($u$)')
    ax1.axhline(0, color='black')
    ax1.axvline(0, color='black')
    ax1.grid(True)
    ax1.legend(loc='upper right')

    ax2.scatter(df_com['b'], df_com['u_complex'], color='blue', s=1 , label='Imaginary part Complex solutions')
    if mark_point is not None:
      ax2.scatter(mark_point[0], mark_point[2], color='green', s=10 , label=mark_point[3])
    ax2.set_title('Complex part plot')
    ax2.set_xlabel('$\\hat{\\beta}$')
    ax2.set_ylabel('Imag($u$)')
    ax2.axhline(0, color='black')
    ax2.axvline(0, color='black')
    ax2.grid(True)
    ax2.legend(loc='upper right')

    fig.suptitle(f'Bifurcation Diagram for $\\hat{{k}} = {klabel}$', fontsize=16)

    plt.tight_layout()

    plt.show()
  except Exception as e:
    print(f'Plot Error: {e}')



def plot_fixed_b(df,b = 0,blabel = None, mark_point = None):
  assert b in df['b'].values
  try:
    if blabel is None:
      blabel = str(b)
    filtered_df = df[df['b']==b]
    df_real = filtered_df[filtered_df['u_complex']==0]
    df_com = filtered_df[filtered_df['u_complex']!=0]

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

    ax1.scatter(df_real['k'], df_real['u_real'], color='red', s=1 , label='Real solutions')
    ax1.scatter(df_com['k'], df_com['u_real'], color='blue', s=1 , label='Real part (Complex solution)')
    if mark_point is not None:
      ax1.scatter(mark_point[0], mark_point[1], color='green', s=10 , label=mark_point[3])
    ax1.set_title('Real part plot')
    ax1.set_xlabel('$\\hat{k}$')
    ax1.set_ylabel('Re($u$)')
    ax1.axhline(0, color='black')
    ax1.axvline(0, color='black')
    ax1.grid(True)
    ax1.legend(loc='upper right')

    ax2.scatter(df_com['k'], df_com['u_complex'], color='blue', s=1 , label='Imaginary part Complex solutions')
    if mark_point is not None:
      ax2.scatter(mark_point[0], mark_point[2], color='green', s=10 , label=mark_point[3])
    ax2.set_title('Complex part plot')
    ax2.set_xlabel('$\\hat{k}$')
    ax2.set_ylabel('Imag($u$)')
    ax2.axhline(0, color='black')
    ax2.axvline(0, color='black')
    ax2.grid(True)
    ax2.legend(loc='upper right')

    fig.suptitle(f'Bifurcation Diagram for $\\hat{{\\beta}} = {blabel}$', fontsize=16)

    plt.tight_layout()

    plt.show()
  except Exception as e:
    print(f'Plot Error: {e}')
```

## Graficos Tridimensionales

Para este contexto se diseñara una única función que recibirá una estructura que contenga los datos de la solución para un conjunto de parametros, para posteriormente realizar un diagrama tridicmensional interactivo en el que se pueden apreciar las tres soluciones coexistiendo en el plano complejo

```python
import plotly.graph_objects as go
import numpy as np

# Generate sample data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Create a 3D surface plot
fig = go.Figure(data=[go.Surface(z=z, x=x, y=y, colorscale='Viridis')])

# Update layout for interactivity
fig.update_layout(
    title='Interactive 3D Surface Plot',
    scene=dict(
        xaxis_title='X Axis',
        yaxis_title='Y Axis',
        zaxis_title='Z Axis'
    )
)

# Show the plot
fig.show()
```
