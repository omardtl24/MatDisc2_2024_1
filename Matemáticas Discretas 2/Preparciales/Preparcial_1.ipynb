### 1. Determine y justifique si $S = \{1, 2, 3, 6, 18\}$ bajo la operación $*$ definida como $a * b = ab$ es un grupo.

Para determinar que el conjunto $S$ sobre la operación $*$ es un grupo, debemos determinar en primer lugar que la operación binaria, es decir que:

$$a * b \in S \,\, \forall_{a,b \in S}$$

Para ello, vamos a organizar el resultado de la operación de cada elemento contra otro en una tabla, donde $a$ representa el elemento de la fila y $b$ representa el elemento de la columna.

\begin{array}{|r|c|c|c|c|c|}
    \hline
     S \times S & \textbf{1} & \textbf{2} & \textbf{3} & \textbf{6} & \textbf{18} \\
    \hline
    \textbf{1} & 1 & 2 & 3 & 6 & 18 \\
    \hline
    \textbf{2} & 2 & 4 & 6 & 12 & 36 \\
    \hline
    \textbf{3} & 3 & 6 & 9 & 18 & 54 \\
    \hline
    \textbf{6}  & 6 & 12 & 18 & 36 & 108 \\
    \hline
    \textbf{18}  & 18 & 36 & 54 & 108 & 324 \\
    \hline
\end{array}

Podemos identificar múltiples casos donde el resultado de $a * b \in S$. Aunque nos interesa para la demostración solo mencionar uno:

$$a=3 \wedge b=18 \Longrightarrow a \in S \wedge b \in S \wedge a * b = 54 \notin S$$

Por lo tanto podemos denotar que:

$$\exists_{a,b \in S}  \,\, a*b \notin S$$

Es decir que la operación $*$ sobre el conjunto $S$ no es binaria. Por lo tanto:

$$\therefore (S,*) \textbf{ NO} \text{ es un grupo}$$

### 2. Dado $(G, ∗)$, demuestre que $(G, ∗)$ es abeliano, si y solo sí, $(x ∗ y)^{−1} = x^{−1} ∗ y^{−1} \,\, \forall_{x, y \in G}$

Para poder demostrar que $(G, ∗) \text{ es abeliano } \Longleftrightarrow (x ∗ y)^{−1} = x^{−1} ∗ y^{−1} \,\, \forall_{x, y \in G}$ debemos realizarlo tanto de izquierda a derecha como de derecha a izqueirda.

#### $(G, ∗) \text{ es abeliano } \Longrightarrow (x ∗ y)^{−1} = x^{−1} ∗ y^{−1} \,\, \forall_{x, y \in G}$

Para la demostración de izquierda a derecha, vamos a partir del la definición de inverso:

\begin{array}{rclcr}
    (x ∗ y) * (x ∗ y)^{−1} &=& e & & \text{Definición de inverso} \\
    x^{-1} * (x ∗ y) * (x ∗ y)^{−1} &=& x^{-1} * e & & \text{Operamos a la izquierda por } x^{-1} \\
    (x^{-1} * ((x ∗ y) * (x ∗ y)^{−1})) * y^{-1} &=& (x^{-1} * e) * y^{-1}& & \text{Operamos a la derecha por } y^{-1} \\
    (x^{-1} * (x ∗ y)) * (x ∗ y)^{−1} * y^{-1} &=& (x^{-1} * e) * y^{-1}& & \text{Propiedad asociativa}\\
    ((x^{-1} * x) ∗ y) * (x ∗ y)^{−1} * y^{-1} &=& (x^{-1} * e) * y^{-1}& & \text{Propiedad asociativa}\\
    (e ∗ y) * (x ∗ y)^{−1} * y^{-1} &=& (x^{-1} * e) * y^{-1}& & \text{Propiedad del inverso}\\
    y * (x ∗ y)^{−1} * y^{-1} &=& x^{-1} * y^{-1}& & \text{Elemento Identidad}\\
    (y * (x ∗ y)^{−1}) * y^{-1} &=& x^{-1} * y^{-1}& & \text{Propiedad Asociativa}\\
    ((x ∗ y)^{−1} * y) * y^{-1} &=& x^{-1} * y^{-1}& & \text{Propiedad Conmutativa (Abeliano)}\\
    (x ∗ y)^{−1} * (y * y^{-1}) &=& x^{-1} * y^{-1}& & \text{Propiedad Asociativa}\\
    (x ∗ y)^{−1} * e &=& x^{-1} * y^{-1}& & \text{Propiedad del inverso}\\
    (x ∗ y)^{−1}&=& x^{-1} * y^{-1}& & \text{Elemento Identidad}\\
\end{array}

$$\therefore (G, ∗) \text{ es abeliano } \Longrightarrow (x ∗ y)^{−1} = x^{−1} ∗ y^{−1} \,\, \forall_{x, y \in G}$$

#### $(x ∗ y)^{−1} = x^{−1} ∗ y^{−1} \,\, \forall_{x, y \in G} \Longrightarrow (G, ∗) \text{ es abeliano } $

Para la demostración de derecha a izquierda, vamos a partir del la condición impuesta sobre el inverso:

\begin{array}{rclcr}
    (x ∗ y)^{−1}&=& x^{-1} * y^{-1}& & \text{Precondición del Inverso}\\
    \left((x ∗ y)^{−1}\right)^{-1}&=&\left(x^{-1} * y^{-1}\right)^{-1}& & \text{Aplicamos inverso a ambos lados}\\
    \left((x ∗ y)^{−1}\right)^{-1}&=&\left(y^{-1}\right)^{-1} * \left(x^{-1}\right)^{-1} & & \text{Inverso del producto}\\
    x ∗ y&=& y * x & & \text{Inverso del Inverso}\\
\end{array}

$$ \therefore (x ∗ y)^{−1} = x^{−1} ∗ y^{−1} \,\, \forall_{x, y \in G} \Longrightarrow (G, ∗) \text{ es abeliano } $$

Dado que se cumple tanto de izquierda a derecha como de derecha a izquierda, podemos decir:

$$\therefore (G, ∗) \text{ es abeliano } \Longleftrightarrow (x ∗ y)^{−1} = x^{−1} ∗ y^{−1} \,\, \forall_{x, y \in G}$$

### 3. Dado $G = \langle x \rangle$, y $|G| = 24$. Liste todos los elementos de $G$ de orden $4$.

Dado que nos mencionan que el grupo es generado a partir del elemento $x$ y que tiene cardinalidad 24, podemos escribirlo de la siguiente manera:

$$G = \langle x \rangle = \{x^{n} : n \in \mathbb{Z}_{24}\} = \{e,x,x^2,x^3 , ... , x^{22},x^{23}\}$$

Para determinar los elementos de orden 4, aprovecharemos la definición de orden sabiendo que $o(x)=24$.

$$o(x^{n})=\dfrac{24}{(24,n)}$$

Dado que buscamos los elementos de orden 4, vamos a establecer una relación entre los elementos de dicho orden y el máximo común divisor entre $n$ y $24$.

$$o(x^{n})=4 \Longrightarrow \dfrac{24}{(24,n)} = 4 \Longrightarrow (24,n) = 6$$

Dado que $24 = 6 \cdot 4$ y que $(24,n)=6$, podemos establecer como criterio para $n$:

$$(24,n) = 6 \wedge 24 = 6 \cdot 4 \Longrightarrow 6 \mid n \wedge 4 \nmid n \Longrightarrow \exists_{k \in \mathbb{Z}}\,\, n = 6k \wedge \forall_{k \in \mathbb{Z}} n \neq 4k$$

Los números que cumplen esa condición son los siguientes:

$$N_6 = \{n \in \mathbb{Z}_{24}: 6 \mid n \wedge 4 \nmid n \} = \{6,18\}$$

Podemos concluir:

$$\therefore G' = \{x^{n} \in G : o(x^{n}) = 4 \} = \{x^{6},x^{18}\}$$

### 4. Dado un grupo $G$ que es abeliano, Demuestre que dado $H = \{x ∈ G : x^2 = e\}$ se cumple que $H ≤ G$.

Para determinar si $H \leq G$, debemos determinar que $H$ no es vacío, que es cerrado bajo la operación del grupo y que es cerrado para los inversos.

#### $H \neq \emptyset$ ?

Para garantizar que $H$ no es vacío, nos basta con encontrar algún elemento de $G$ que pertenezca a $H$. Para ello, vamos a aprovechar el elemento identidad.

$$e \in G \,\, \Longrightarrow  \,\, xe = ex = e \,\, \forall_{x \in G} \,\, \Longrightarrow \,\, ee = e \,\, \Longrightarrow \,\, e^{2}=e \,\, \Longrightarrow \,\, e \in H$$

Dado que hemos encontrado un elemento que pertenece a $H$, podemos concluir que:

$$\therefore H \neq \emptyset$$

#### $\forall_{x,y \in H} \,\, xy \in H$ ?

Debemos demostrar lo siguiente:

$$a \in H \,\, \wedge \,\, b \in H \,\, \overset{?}{\Longrightarrow} \,\, ab \in H$$

Para saber si $ab \in H$, debemos probar la propiedad asociada a ese subgrupo. En este caso seria que $(ab)^2 = e$

\begin{array}{rclcr}
   (ab)^2 & = & (ab)(ab) && \text{Separamos en notación multiplicativa}\\
          & = & (ab)(ba) && \text{Propiedad Conmutativa (Abeliano)}\\
          & = & ((ab)b)a && \text{Propiedad Asociativa}\\
          & = & (a(bb))a && \text{Propiedad Asociativa}\\
          & = & (ae)a && bb=b^{2}=e\\
          & = & aa && \text{Elemento Identidad}\\
          & = & e && aa=a^{2}=e\\
\end{array}

Dado que demostramos que $ab$ cumple con la propiedad definida en el subgrupo $H$, podemos concluir que:

$$\therefore \forall_{x,y \in H} \,\, xy \in H$$

#### $\forall_{x \in H} \,\, x^{-1} \in H$ ?

Para esta demostración, vamos a aprovechar la definición de elemento inverso que es la siguiente:

$$\forall_{x \in G}\,\, \exists_{x^{-1} \in G} \,\, xx^{-1} = x^{-1}x = e$$

En nuestro caso, vamos a visualizar la propiedad que nos da $H$

$$\forall_{x \in H} x^{2} = e$$

No hay que ser muy metódicos para visualizar que x es el inverso de si mismo; y por consiguiente, ya se encuentra en $H$.

$$\forall_{x \in H} \,\, x^2 = e \,\, \wedge \,\, xx^{-1} = x^{-1}x = e \,\, \Longrightarrow \,\, x^{-1} = x \,\, \Longrightarrow \,\, x^{-1} \in H$$

Esto nos permite concluir que:

$$\forall_{x \in H} \,\, x^{-1} \in H$$

#### Finalmente

Dado que pudimos demostrar que $H$ es no vacío, cerrado y cerrado para los inversos, podemos concluir que:

$$H \leq G$$

### 6. Encuentre el orden de $(7, 3, 4)$ en $\mathbb{Z}_{18} \times \mathbb{Z}_{5} \times \mathbb{Z}_{8}$.

Para determinar el orden de $(7, 3, 4)$ en $\mathbb{Z}_{18} \times \mathbb{Z}_{5} \times \mathbb{Z}_{8}$, debemos identificar el orden de cada elemento en su grupo de partida. Para ello, aprovecharemos la relación entre el orden del generador y su máximo común divisor con el elemento.

El orden de $7$ en $\mathbb{Z}_{18}$ es:

$$o(7) = \dfrac{18}{(18,7)} = \dfrac{18}{1} = 18$$

El orden de $3$ en $\mathbb{Z}_{5}$ es:

$$o(3) = \dfrac{5}{(5,3)} = \dfrac{5}{1} = 5$$

El orden de $4$ en $\mathbb{Z}_{8}$ es:

$$o(4) = \dfrac{8}{(8,4)} = \dfrac{8}{4} = 2$$

Ya conociendo el orden de cada elemento, aprovecharemos la relación de estos y el mínimo común múltiplo para determinar el orden de la tupla.

$$o((7,3,4))=\text{mcm}(o(7),o(3),o(4))=\text{mcm}(18,5,2)=90$$

Podemos concluir que:

$$\therefore o((7,3,4))=90$$

### 7. Determine si la siguiente función es inyectiva, sobreyectiva, total, biyectiva. $$ \begin{array}{rcrcl} h & : & \mathbb{N} & \rightarrow & \mathbb{N} \\ & & n & \mapsto & \dfrac{n^2 - 1}{n-1} \end{array} $$

#### Es Total?

Para determinar que la función es total, se debe cumplir lo siguiente:

$$\forall_{a \in A} \,\, \exists_{b \in B} \,\, f(a)=b$$

Debemos determinar si la función está definida para cualquier número natural en el conjunto de salida. En el caso de la función dada, podemos encontrar una indeterminación cuando el denominador sea igual a 0.

$$h(n) = \dfrac{n^2 - 1}{n-1} \Longrightarrow n-1 \neq 0 \Longrightarrow n \neq 1$$

Dado que la función no está definida para $n=1$ y $1 \in \mathbb{N}$, se puede concluir que:

$$\exists_{n \in \mathbb{N}} \,\, \forall_{n' \in \mathbb{N}} \,\, h(n) \neq n'$$

Esto nos permite decir que:

$$\therefore h(n) \text{ NO es una función total}$$

#### Es Inyectiva?

Para determinar que la función es inyectiva, debemos verificar si cumple:

$$f(x)=z \wedge f(y)=z \Longrightarrow x=y$$

Vamos a aprovechar las propiedades del álgebra y la aritmética para tratar de probar lo siguiente:

 $$\dfrac{x^2 - 1}{x-1} = z  \wedge \dfrac{y^2 - 1}{y-1}=z \overset{?}{\Longrightarrow} x=y$$

Vamos a tener en cuenta que la función no está definida para $n=1$

\begin{array}{rclcr}
    \dfrac{x^2 - 1}{x-1} &=& \dfrac{y^2 - 1}{y-1}&& \text{Partimos de nuestra precondición }h(x)=h(y) \\ \\
    \dfrac{(x+1)(x-1)}{x-1} &=& \dfrac{(y+1)(y-1)}{y-1}&& \text{Aplicamos diferencia de cuadrados}\\ \\
    x+1 &=& y+1 && \text{Simplificamos dado que }x \neq 1 \wedge y\neq 1 \\ \\
    x &=& y && \\ \\
\end{array}

Podemos concluir que:

$$\therefore h(n) \text{ SI es una función inyectiva}$$

#### Es Sobreyectiva?

Para determinar que la función es sobreyectiva, debemos verificar si cumple:

$$\forall_{b \in B} \,\, \exists_{a \in A} \,\, f(a)=b$$

En este caso, por inspección vamos a identificar un elemento del conjunto de llegada que no permita que se cumpla la condición. Definamos $b=2$ como elemento del conjunto de llegada.

$$f(n)=2 \Longrightarrow \dfrac{n^2 - 1}{n-1} = 2$$

Usaremos propiedades aritméticas y algebraicas para determinar si existe un elemento $n$ del conjunto de salida que mapee a $2$

\begin{array}{rclcr}
    \dfrac{n^2 - 1}{n-1} &=& 2&& \text{Punto de Partida} \\ \\
    n^2 - 1 &=& 2n - 2&& \text{Despejamos bajo la condición de que }n\neq1\\ \\
    n^2 - 2n + 1 &=& 0&& \\ \\
    (n-1)^{2} &=& 0&& \\ \\
    n-1 &=& 0&& \\ \\
    n &=& 1&& \text{Se ha llegado a una contradicción}\\ \\
\end{array}

Hemos identificado que el único candidato posible para mapear a $2$ es $n=1$. Sin embargo, anteriormente concluimos que la función no estaba definida para $n=1$. Lo que nos permite concluir que:

$$\exists_{n' \in \mathbb{N}} \,\, \forall_{n \in \mathbb{N}} \,\, h(n)\neq n'$$

Es decir:

$$\therefore h(n) \text{ NO es una función sobreyectiva}$$

#### Es Biyectiva?

Para que la función sea biyectiva, debe cumplirse que sea inyectiva, sobreyectiva y total. Dado que la función no es sobreyectiva ni es total, podemos decir:

$$\therefore h(n) \text{ NO es una función biyectiva}$$

$$
