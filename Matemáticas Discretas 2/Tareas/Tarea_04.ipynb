### 1. Considerando el grupo de enteros $G=(\mathbb{Z}_4,\oplus)$ encuentre el orden de cada elemento de $\mathbb{Z}_4=\{0,1,2,3\}$


Dado que nuestra opración a analizar es la adición módulo 4, podremos escribir la n-ésima operación de un elemento $x$ consigo mismo de la siguiente forma:

\begin{array}{ccccc}
    x^{n} &=&\underbrace{x \oplus x \oplus x ...}  &=& n\cdot x\mod 4\\
    && n\text{-veces} && \\
\end{array}

Adicionalmente, podemos denotar que el elemento identidad sobre nuestra operación es el 0 $(e = 0)$, ya que:

$$\forall_{x \in \mathbb{Z}_4} \,\, x \oplus 0 =  0 \oplus x = x \mod 4$$

El siguiente paso, es describir el orden de cada uno de los elementos de nuestro conjunto. Para esto, vamos a aprovechar la definición de orden y la definición del módulo:

$$o(x) = n \Longleftrightarrow x^{n} = e$$

$$x \mod n = r \Longleftrightarrow \exists_{k \in \mathbb{Z}} \,\, k\cdot n + r = x$$

Para hallar el orden de cada elemento de nuestro conjunto, podemos aplicar lo siguiente:

\begin{array}{rcl}
    o(x) = n &\Longleftrightarrow& x^{n} = 0 \\
    & & n\cdot x\mod 4 = 0 \\
    & & k\cdot 4 + 0 = n \cdot x \\
    & & \dfrac{4 \cdot k}{x} = n \\
\end{array}

Se puede concluir que para todo elemento $x \in \mathbb{Z}_4$, su orden se puede describir como el número entero positivo más pequeño que cumpla $\dfrac{4 \cdot k}{x} \in \mathbb{Z}^{+} \wedge k \in  \mathbb{Z}$

#### **Para $x = 0$:**

Dado que el número cero es el elemento identidad, podemos denotar que:

$$0^{1} = 0 \mod 4 = 0$$

Dado que 1 es el entero positivo más pequeño que permite realizar la operación, se puede concluir que:

$$o(0)=1$$

#### **Para $x = 1$:**

A partir de la fórmula planteada $\dfrac{4 \cdot k}{1} = 4 \cdot k$, podemos denotar lo siguiente:

$$k=1 \Longrightarrow 4 \cdot k  \in \mathbb{Z}^{+} \wedge k \in  \mathbb{Z} $$

Dado que el número entero más pequeño que permite cumplir la condición es $k=1$, se puede concluir que:

$$o(1)=\dfrac{4 \cdot 1}{1} = 4$$

#### **Para $x = 2$:**

A partir de la fórmula planteada $\dfrac{4 \cdot k}{2} = 2 \cdot k$, podemos denotar lo siguiente:

$$k=1 \Longrightarrow 2 \cdot k  \in \mathbb{Z}^{+} \wedge k \in  \mathbb{Z} $$

Dado que el número entero más pequeño que permite cumplir la condición es $k=1$, se puede concluir que:

$$o(2)=\dfrac{2 \cdot 1}{1} = 2$$

#### **Para $x = 3$:**

A partir de la fórmula planteada $\dfrac{4 \cdot k}{3}$, podemos denotar lo siguiente:

$$k=3 \Longrightarrow \dfrac{4 \cdot k}{3}  \in \mathbb{Z}^{+} \wedge k \in  \mathbb{Z} $$

Dado que el número entero más pequeño que permite cumplir la condición es $k=3$, se puede concluir que:

$$o(3)=\dfrac{4 \cdot 3}{3} = 4$$


#### Finalmente:


\begin{array}{lcl}

\therefore & o(0) = 1 \\
           & o(1) = 4 \\
           & o(2) = 2 \\
           & o(3) = 4 \\
\end{array}

### 3. Encuentre el orden de 1 en  $G=(\mathbb{Z},+)$

Para determinar el orden de $x=1$ en el grupo $G$, es necesario determinar el elemento identidad de este grupo. Podemos visualizar que $e=0$ es el elemento identidad ya que:

$$\forall_{x \in \mathbb{Z}} \,\,\, x + 0  = 0 + x = x$$

Adicionalmente, introduciremos la siguiente nomenclatura para hablar de la aplicación de  la n-ésima operación sobre un elemento:

\begin{array}{ccccc}
    x^{n} &=&\underbrace{x + x + x ...}  &=& n\cdot x\\
    && n\text{-veces} && \\
\end{array}

Para determinar el orden de $x=1$, debemos encontrar el mínimo n tal que:

$$1^{n}=0 \Longrightarrow n \cdot 1 = 0 , n \in \mathbb{Z}^{+}$$

Podemos denotar que el único valor de $n$ que cumple con nuestro problema es $n=0$. Sin embargo, $0 \notin \mathbb{Z}^{+}$, por lo que **NO** existe un valor de $n$ válido para nuestro problema. A partir de eso, se puede concluir que:

$$\therefore o(1)=\infty$$

### 4. ¿Es $G=(\mathbb{Z}_4,\oplus)$ un grupo cíclico? Si lo es, ¿cuáles son sus generadores?

Para determinar que $G$ es un grupo cíclico debemos determinar si existe un elemento tal que

$$\langle x \rangle = \{nx \mod 5 : n\in \mathbb{Z}\} = \mathbb{Z}_4$$

No es necesario ser muy metódicos para denotar que $x=1$ es un valor que cumple esta propiedad, solo es necesario evaluarlo en los 4 casos más simples para hacerlo visible:

\begin{array}{cccc}
    1^{1} &=& 1 \mod 4 &=& 1\\ 
    1^{2} &=& 2 \mod 4 &=& 2\\ 
    1^{3} &=& 3 \mod 4 &=& 3\\ 
    1^{4} &=& 4 \mod 4 &=& 0\\  
\end{array}

$$\langle 1 \rangle = \{0,1,2,3\} = \mathbb{Z}_4$$

Con este proceso, hemos denotado que $G$ es un grupo cíclico. El siguiente paso es determinar todos sus posibles generadores. Para ello, vamos a apoyarnos de la siguiente propiedad:

$$G = \langle x \rangle \Longrightarrow o(x) = |G|$$

Y su respectiva contrarreciproca:

$$o(x) \neq |G|  \Longrightarrow G \neq \langle x \rangle$$

A partir de esto, podemos descartar y determinar posibles candidatos para ser generadores de $G$. A partir de este criterio, podemos descartar $x = 0 \wedge x = 2$; y tendriamos de candidatos a $x = 1 \wedge x = 3$

Con el procedimiento anterior ya demostramos que $x=1$ es un generador de $G$. El siguiente paso es realizar el mismo proceso para $x=3$

\begin{array}{cccc}
    3^{1} &=& 3 \mod 4 &=& 3\\ 
    3^{2} &=& 6 \mod 4 &=& 2\\ 
    3^{3} &=& 9 \mod 4 &=& 1\\ 
    3^{4} &=& 12 \mod 4 &=& 0\\ 
\end{array}

$$\langle 3 \rangle = \{0,1,2,3\} = \mathbb{Z}_4$$

Con ese proceso, se puede concluir que $x=3$ también es un generador de $G$.

\begin{array}{lcl}

\therefore & G\text{ es un grupo cíclico} \\
           & \text{Gen}(G) = \{1,3\}
\end{array}
